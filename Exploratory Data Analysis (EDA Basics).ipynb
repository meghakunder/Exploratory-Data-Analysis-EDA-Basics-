**Q1. What is the average, median, and standard deviation of house prices?**

```python
print(df['Price'].agg(['mean', 'median', 'std']))
```

The average house price is approximately 538,932, while the median price is 450,000. Since the mean is higher than the median, it indicates that some high-priced houses are increasing the average price. The standard deviation is 367,532, which shows that house prices vary significantly across the dataset.

**Q2. Which number of bedrooms is most common?**

```python
print(df['number of bedrooms'].mode()[0])
```

The mode of the number of bedrooms column shows the most common bedroom count in the dataset. This indicates the type of houses that are most frequently available or demanded in the market.

**Q3. Check if the price data is skewed.**

```python
print(df['Price'].skew())
```

The skewness value of the Price column is 4.27, which indicates that the data is highly positively skewed (right-skewed). This means most houses have lower to moderate prices, while a few houses have extremely high prices that pull the distribution toward the right side.

**Q4. What is the average price for each number of bedrooms?**

```python
df.groupby('number of bedrooms')['Price'].mean()
```

**Interpretation**
1. Average house price increases as the number of bedrooms increases.

2. Houses with 1–3 bedrooms have comparatively lower average prices.

3. Houses with 7–8 bedrooms have the highest average prices, exceeding 1 million.

4. There is a positive relationship between number of bedrooms and house price.

5. Bedroom categories such as 9, 10, 11, and 33 show irregular average prices, possibly due to outliers or fewer observations in the dataset.


**Q5.  What is the relationship between living area and price? (Use : Area without Basement).**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Area of the house(excluding basement)', y='Price', data=df)

plt.title("Living Area vs Price")

plt.show()
```

```python
print(df[['Area of the house(excluding basement)','Price']].corr())
```

1. The correlation between living area (excluding basement) and price is **0.615**.

2. This indicates a moderately strong positive relationship between the two variables.

3. As the living area increases, house prices also tend to increase.

4. Houses with larger usable living space generally have higher market value.

5. Living area is an important factor influencing house prices.


**Q6. Identify any anomalies where houses have high prices but low areas.**

```python
anomalies = df[
    (df['Price'] > df['Price'].quantile(0.75)) &
    (df['Area of the house(excluding basement)'] < 
     df['Area of the house(excluding basement)'].quantile(0.25))
]

print(anomalies[['Price', 'Area of the house(excluding basement)']])
```


1. Some houses have unusually high prices despite smaller living areas.

2. These properties can be considered anomalies or outliers in the dataset.

3. High prices for smaller houses may be influenced by factors such as location, luxury features, amenities, or market demand.

4. These anomalies indicate that house price is not determined only by living area.

5. Other factors also significantly influence property valuation.


**Q7. Compare average price based on number of floors and houses with or without waterfront.**

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
print(
    df.groupby(
        ['number of floors', 'waterfront present']
    )['Price'].mean()
)
```

1. Houses with waterfront access have higher average prices than houses without waterfront access.

2. Average house price generally increases with the number of floors.

3. 2.5-floor waterfront houses have the highest average prices.

4. Multi-floor waterfront houses are premium properties in the dataset.

5. Both waterfront access and number of floors positively influence house prices.

**Q9. Which location (zipcode/area) has the highest average price?**

```python
df.groupby('Postal Code')['Price'].mean().nlargest(1)
```

**Interpretation**
